```python
BHD = pd.read_csv("/content/drive/MyDrive/IRIS/Bengaluru_House_Data.csv")
BHD
```

```python
BHD.isna().sum()
```

```python
for column in BHD.columns:
    if BHD[column].dtype == 'object':  # Categorical
        BHD[column].fillna('Unknown', inplace=True)
    else:  # Numerical
        BHD[column].fillna(BHD[column].median(), inplace=True)
```

```python
BHD.dropna(inplace=True)
```

```python
BHD['society'].fillna('Unknown', inplace=True)  # Example for one column
BHD['bath'].fillna(BHD['bath'].median(), inplace=True)  # For numerical
BHD['size'].fillna(BHD['size'].mode()[0], inplace=True)  # For categorical
```

```python
print(BHD.isnull().sum())  # Shows remaining missing values column-wise
```

```python
import re

# Extract numerical values from the 'size' column
BHD['size'] = BHD['size'].str.extract(r'(\d+)', expand=False)

# Convert the extracted values to integers
BHD['size'] = BHD['size']

# Verify the changes
print(BHD['size'].head())
print(BHD['size'].dtypes)  # Check the data type
```

```python

# Define a function to process total_sqft
def process_sqft(value):
    try:
        # Check if the value contains a range
        if '-' in value:
            values = list(map(float, value.split('-')))
            return sum(values) / 2  # Return the average of the range
        # If it's a single value, return it as a float
        return float(value)
    except ValueError:
        # Handle cases with non-numeric values
        return None

# Apply the function to the total_sqft column
BHD['total_sqft'] = BHD['total_sqft'].apply(process_sqft)

# Convert the column to numeric type
BHD['total_sqft'] = pd.to_numeric(BHD['total_sqft'], errors='coerce')

# Verify the changes
print(BHD['total_sqft'].head())
```
